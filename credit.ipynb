```python

import numpy as np 
import pandas as pd 
import tensorflow as tf 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split, StratifiedKFold 
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (
    confusion_matrix, classification_report, roc_auc_score, 
    average_precision_score, RocCurveDisplay, PrecisionRecallDisplay
)
import zipfile
import os

np.random.seed(42) 
tf.random.set_seed(42)

print("="*80)
print("CREDIT CARD FRAUD DETECTION - DEEP NEURAL NETWORK")
print("="*80)
print(f"TensorFlow Version: {tf.__version__}")
print(f"NumPy Version: {np.__version__}")
print(f"Pandas Version: {pd.__version__}")

print("\n" + "="*80)
print("STEP 1: LOADING DATASET")
print("="*80)
zip_path = r"C:\Users\Janane S B\Downloads\creditcard.csv.zip"
print(f"\nExtracting CSV from ZIP file...")
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    file_list = zip_ref.namelist()
    print(f"Files in ZIP: {file_list}")
    csv_file = [f for f in file_list if f.endswith('.csv')][0]
    print(f"Reading: {csv_file}")
   
    with zip_ref.open(csv_file) as f:
        df = pd.read_csv(f)

print(f"\nDataset loaded successfully!")
print(f"Dataset Shape: {df.shape}")
print(f"\nColumn Names:")
print(df.columns.tolist())
print(f"\nData Types:")
print(df.dtypes)
print(f"\nFirst few rows:")
print(df.head())

print("\n" + "="*80)
print("STEP 2: DATA PREPROCESSING")
print("="*80)

possible_target_names = ['Class', 'class', 'CLASS', 'label', 'Label', 'target', 'Target', 'fraud', 'Fraud']
target_column = None

for col_name in possible_target_names:
    if col_name in df.columns:
        target_column = col_name
        break

if target_column is None:
    for col in df.columns:
        if 'class' in col.lower() or 'fraud' in col.lower():
            target_column = col
            break

if target_column is None:
    target_column = df.columns[-1]
    print(f"\n⚠️ WARNING: Could not find standard target column name.")
    print(f"   Assuming last column '{target_column}' is the target variable.")

print(f"\n✓ Using '{target_column}' as target column")

X = df.drop(target_column, axis=1).copy()
y = df[target_column].copy()

print(f"\nOriginal Features shape: {X.shape}")
print(f"Target shape: {y.shape}")

non_numeric_cols = X.select_dtypes(include=['object']).columns.tolist()

if len(non_numeric_cols) > 0:
    print(f"\n⚠️ Found {len(non_numeric_cols)} non-numeric columns: {non_numeric_cols}")
    print("Encoding categorical variables...")
    label_encoders = {}
    for col in non_numeric_cols:
        print(f"  Encoding column: {col}")
        print(f"    Unique values: {X[col].nunique()}")
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col].astype(str))
        label_encoders[col] = le
    
    print("✓ Categorical encoding completed")
if y.dtype == 'object':
    print(f"\n⚠️ Target column '{target_column}' is non-numeric")
    print(f"Unique values: {y.unique()}")
    print("Encoding target variable...")
    le_target = LabelEncoder()
    y = le_target.fit_transform(y)
    print(f"Encoded classes: {le_target.classes_}")
    print("✓ Target encoding completed")
unique_targets = np.unique(y)
if len(unique_targets) != 2:
    print(f"\n⚠️ WARNING: Target has {len(unique_targets)} unique values: {unique_targets}")
    print("This code is designed for binary classification.")
    if len(unique_targets) > 2:
        print("Converting to binary: 0 vs all others")
        y = (y == 0).astype(int)

print(f"\n{target_column} Distribution:")
print(pd.Series(y).value_counts())
print(f"\nClass 1 Percentage: {(y.sum() / len(y)) * 100:.4f}%")

print(f"\nFinal Features shape: {X.shape}")
print(f"Final Target shape: {y.shape}")
print(f"\nFeature data types:")
print(X.dtypes.value_counts())
X = X.values
y = y if isinstance(y, np.ndarray) else y.values

print(f"\n✓ Data preprocessing completed")
print(f"X shape: {X.shape}, dtype: {X.dtype}")
print(f"y shape: {y.shape}, dtype: {y.dtype}")

def preprocess_data(X_train, X_test): 
    """Standardize features using StandardScaler"""
    scaler = StandardScaler() 
    X_train_scaled = scaler.fit_transform(X_train) 
    X_test_scaled = scaler.transform(X_test) 
    return X_train_scaled, X_test_scaled 

def compute_class_weights(y): 
    """Compute balanced class weights for imbalanced dataset"""
    neg, pos = np.bincount(y) 
    total = neg + pos 
    weights = {0: total / (2 * neg), 1: total / (2 * pos)}
    print(f"\nClass Weights: Class 0={weights[0]:.4f}, Class 1={weights[1]:.4f}")
    print(f"Weight Ratio (Class 1/Class 0): {weights[1]/weights[0]:.2f}")
    return weights

def build_dnn(input_dim): 
    """Build Deep Neural Network with Batch Normalization and Dropout"""
    model = tf.keras.Sequential([ 
        tf.keras.layers.Input(shape=(input_dim,)), 
        tf.keras.layers.Dense(128, activation="relu"), 
        tf.keras.layers.BatchNormalization(), 
        tf.keras.layers.Dropout(0.4), 
        tf.keras.layers.Dense(64, activation="relu"), 
        tf.keras.layers.BatchNormalization(), 
        tf.keras.layers.Dropout(0.3), 
        tf.keras.layers.Dense(32, activation="relu"), 
        tf.keras.layers.BatchNormalization(), 
        tf.keras.layers.Dropout(0.2), 
        tf.keras.layers.Dense(16, activation="relu"), 
        tf.keras.layers.Dense(1, activation="sigmoid") 
    ]) 
    
    model.compile( 
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3), 
        loss="binary_crossentropy", 
        metrics=[ 
            tf.keras.metrics.AUC(name="roc_auc"), 
            tf.keras.metrics.AUC(name="pr_auc", curve="PR") 
        ] 
    ) 
    return model 

def evaluate_model(model, X_test, y_test, exp_name=""): 
    """Evaluate model and display comprehensive metrics"""
    y_pred_prob = model.predict(X_test, verbose=0).ravel() 
    y_pred = (y_pred_prob >= 0.5).astype(int) 
    
    print(f"\n{'='*80}")
    print(f"{exp_name} - EVALUATION METRICS")
    print(f"{'='*80}")
    
    print("\nConfusion Matrix:") 
    cm = confusion_matrix(y_test, y_pred)
    print(cm) 
    
    print("\nClassification Report:") 
    print(classification_report(y_test, y_pred, digits=4, target_names=['Class 0', 'Class 1']))
    
    roc_auc = roc_auc_score(y_test, y_pred_prob) 
    pr_auc = average_precision_score(y_test, y_pred_prob) 
    
    print(f"\nPerformance Scores:")
    print(f"  ROC-AUC: {roc_auc:.4f}") 
    print(f"  PR-AUC:  {pr_auc:.4f}") 
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
   
    RocCurveDisplay.from_predictions(y_test, y_pred_prob, ax=axes[0])
    axes[0].set_title(f'{exp_name} - ROC Curve (AUC={roc_auc:.4f})', fontweight='bold')
    axes[0].grid(True, alpha=0.3)
    
    PrecisionRecallDisplay.from_predictions(y_test, y_pred_prob, ax=axes[1])
    axes[1].set_title(f'{exp_name} - PR Curve (AUC={pr_auc:.4f})', fontweight='bold')
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show() 
    
    return roc_auc, pr_auc 
# EXPERIMENT 1: 80-20 TRAIN-TEST SPLIT

print("\n" + "="*80)
print("EXPERIMENT 1: 80-20 TRAIN-TEST SPLIT")
print("="*80)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
print(f"Train size: {len(X_train)}, Test size: {len(X_test)}")

X_train_scaled, X_test_scaled = preprocess_data(X_train, X_test)
model_exp1 = build_dnn(X_train_scaled.shape[1])
class_weights = compute_class_weights(y_train)

print("\nModel Architecture:")
model_exp1.summary()

print("\nTraining model...")
history_exp1 = model_exp1.fit(
    X_train_scaled, y_train, 
    epochs=20, 
    batch_size=2048, 
    class_weight=class_weights, 
    verbose=1
)

plt.figure(figsize=(10, 5))
plt.plot(history_exp1.history["loss"], label="Train Loss", linewidth=2)
plt.title("Experiment 1 - Training Loss Curve", fontsize=14, fontweight='bold')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

roc_exp1, pr_exp1 = evaluate_model(model_exp1, X_test_scaled, y_test, exp_name="Experiment 1")
# EXPERIMENT 2: 70-30 TRAIN-TEST SPLIT

print("\n" + "="*80)
print("EXPERIMENT 2: 70-30 TRAIN-TEST SPLIT")
print("="*80)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)
print(f"Train size: {len(X_train)}, Test size: {len(X_test)}")

X_train_scaled, X_test_scaled = preprocess_data(X_train, X_test)
model_exp2 = build_dnn(X_train_scaled.shape[1])
class_weights = compute_class_weights(y_train)

print("\nTraining model...")
history_exp2 = model_exp2.fit(
    X_train_scaled, y_train, 
    epochs=20, 
    batch_size=2048, 
    class_weight=class_weights, 
    verbose=1
)

plt.figure(figsize=(10, 5))
plt.plot(history_exp2.history["loss"], label="Train Loss", linewidth=2)
plt.title("Experiment 2 - Training Loss Curve", fontsize=14, fontweight='bold')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

roc_exp2, pr_exp2 = evaluate_model(model_exp2, X_test_scaled, y_test, exp_name="Experiment 2")
# EXPERIMENT 3: TRAIN-VALIDATION-TEST SPLIT WITH EARLY STOPPING

print("\n" + "="*80)
print("EXPERIMENT 3: TRAIN-VALIDATION-TEST SPLIT WITH EARLY STOPPING")
print("="*80)

# 60% train, 20% validation, 20% test
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.25, stratify=y_temp, random_state=42
)
print(f"Train size: {len(X_train)}, Validation size: {len(X_val)}, Test size: {len(X_test)}")

X_train_scaled, X_val_scaled = preprocess_data(X_train, X_val)
X_test_scaled = StandardScaler().fit_transform(X_test)

model_exp3 = build_dnn(X_train_scaled.shape[1])
class_weights = compute_class_weights(y_train)

early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_pr_auc", 
    mode="max", 
    patience=5, 
    restore_best_weights=True,
    verbose=1
)

print("\nTraining model with early stopping...")
history_exp3 = model_exp3.fit(
    X_train_scaled, y_train, 
    validation_data=(X_val_scaled, y_val), 
    epochs=50, 
    batch_size=2048, 
    class_weight=class_weights, 
    callbacks=[early_stop], 
    verbose=1
)
plt.figure(figsize=(10, 5))
plt.plot(history_exp3.history["loss"], label="Train Loss", linewidth=2)
plt.plot(history_exp3.history["val_loss"], label="Validation Loss", linewidth=2)
plt.title("Experiment 3 - Training & Validation Loss Curves", fontsize=14, fontweight='bold')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

roc_exp3, pr_exp3 = evaluate_model(model_exp3, X_test_scaled, y_test, exp_name="Experiment 3")
# EXPERIMENT 4: REDUCED TEST SIZE WITH EARLY STOPPING

print("\n" + "="*80)
print("EXPERIMENT 4: REDUCED TEST SIZE WITH EARLY STOPPING")
print("="*80)

# 70% train, 15% validation, 15% test
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, stratify=y, random_state=42
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.1765, stratify=y_temp, random_state=42
)
print(f"Train size: {len(X_train)}, Validation size: {len(X_val)}, Test size: {len(X_test)}")

X_train_scaled, X_val_scaled = preprocess_data(X_train, X_val)
X_test_scaled = StandardScaler().fit_transform(X_test)

model_exp4 = build_dnn(X_train_scaled.shape[1])
class_weights = compute_class_weights(y_train)

early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_pr_auc", 
    mode="max", 
    patience=5, 
    restore_best_weights=True,
    verbose=1
)

print("\nTraining model with early stopping...")
history_exp4 = model_exp4.fit(
    X_train_scaled, y_train, 
    validation_data=(X_val_scaled, y_val), 
    epochs=50, 
    batch_size=2048, 
    class_weight=class_weights, 
    callbacks=[early_stop], 
    verbose=1
)

plt.figure(figsize=(10, 5))
plt.plot(history_exp4.history["loss"], label="Train Loss", linewidth=2)
plt.plot(history_exp4.history["val_loss"], label="Validation Loss", linewidth=2)
plt.title("Experiment 4 - Training & Validation Loss Curves", fontsize=14, fontweight='bold')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

roc_exp4, pr_exp4 = evaluate_model(model_exp4, X_test_scaled, y_test, exp_name="Experiment 4")
# EXPERIMENT 5: STRATIFIED 5-FOLD CROSS VALIDATION

print("\n" + "="*80)
print("EXPERIMENT 5: STRATIFIED 5-FOLD CROSS VALIDATION")
print("="*80)

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
roc_scores = [] 
pr_scores = [] 

for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), 1): 
    print(f"\n{'-'*80}")
    print(f"FOLD {fold}/5")
    print(f"{'-'*80}")
    
    X_train_fold = X[train_idx]
    X_test_fold = X[test_idx]
    y_train_fold = y[train_idx]
    y_test_fold = y[test_idx]
    
    X_train_scaled, X_test_scaled = preprocess_data(X_train_fold, X_test_fold)
    
    model_fold = build_dnn(X_train_scaled.shape[1])
    class_weights = compute_class_weights(y_train_fold)
    
    print(f"\nTraining Fold {fold}...")
    history_fold = model_fold.fit(
        X_train_scaled, y_train_fold, 
        epochs=20, 
        batch_size=2048, 
        class_weight=class_weights, 
        verbose=0
    )
    
    plt.figure(figsize=(10, 5))
    plt.plot(history_fold.history["loss"], label=f"Fold {fold} Train Loss", linewidth=2)
    plt.title(f"Experiment 5 - Fold {fold} Training Loss Curve", fontsize=14, fontweight='bold')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    y_pred_prob = model_fold.predict(X_test_scaled, verbose=0).ravel()
    y_pred = (y_pred_prob >= 0.5).astype(int)
    
    roc = roc_auc_score(y_test_fold, y_pred_prob)
    pr = average_precision_score(y_test_fold, y_pred_prob)
    
    print(f"\nFold {fold} Metrics:")
    print(f"  ROC-AUC: {roc:.4f}")
    print(f"  PR-AUC:  {pr:.4f}")
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    RocCurveDisplay.from_predictions(y_test_fold, y_pred_prob, ax=axes[0])
    axes[0].set_title(f'Fold {fold} - ROC Curve (AUC={roc:.4f})', fontweight='bold')
    axes[0].grid(True, alpha=0.3)
    PrecisionRecallDisplay.from_predictions(y_test_fold, y_pred_prob, ax=axes[1])
    axes[1].set_title(f'Fold {fold} - PR Curve (AUC={pr:.4f})', fontweight='bold')
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    roc_scores.append(roc) 
    pr_scores.append(pr) 

print("\n" + "="*80)
print("EXPERIMENT 5 - CROSS VALIDATION SUMMARY")
print("="*80)
print(f"\nROC-AUC Scores per Fold: {[f'{s:.4f}' for s in roc_scores]}")
print(f"PR-AUC Scores per Fold:  {[f'{s:.4f}' for s in pr_scores]}")
print(f"\nMean ROC-AUC: {np.mean(roc_scores):.4f} ± {np.std(roc_scores):.4f}") 
print(f"Mean PR-AUC:  {np.mean(pr_scores):.4f} ± {np.std(pr_scores):.4f}")
# FINAL COMPARATIVE ANALYSIS

print("\n" + "="*80)
print("COMPREHENSIVE EXPERIMENTAL RESULTS SUMMARY")
print("="*80)

results_summary = pd.DataFrame({
    'Experiment': [
        'Exp 1: 80-20 Split',
        'Exp 2: 70-30 Split',
        'Exp 3: Train-Val-Test + Early Stop',
        'Exp 4: Reduced Test + Early Stop',
        'Exp 5: 5-Fold CV (Mean)'
    ],
    'ROC-AUC': [
        roc_exp1,
        roc_exp2,
        roc_exp3,
        roc_exp4,
        np.mean(roc_scores)
    ],
    'PR-AUC': [
        pr_exp1,
        pr_exp2,
        pr_exp3,
        pr_exp4,
        np.mean(pr_scores)
    ]
})

print("\n" + results_summary.to_string(index=False))
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

experiments = results_summary['Experiment'].values
roc_values = results_summary['ROC-AUC'].values
pr_values = results_summary['PR-AUC'].values

axes[0].bar(range(len(experiments)), roc_values, color='steelblue', alpha=0.7, edgecolor='black')
axes[0].set_xticks(range(len(experiments)))
axes[0].set_xticklabels(experiments, rotation=15, ha='right')
axes[0].set_ylabel('ROC-AUC Score', fontweight='bold')
axes[0].set_title('ROC-AUC Comparison Across Experiments', fontsize=14, fontweight='bold')
axes[0].set_ylim([min(roc_values) - 0.05, 1.0])
axes[0].grid(True, alpha=0.3, axis='y')
for i, v in enumerate(roc_values):
    axes[0].text(i, v + 0.005, f'{v:.4f}', ha='center', va='bottom', fontweight='bold', fontsize=9)

axes[1].bar(range(len(experiments)), pr_values, color='coral', alpha=0.7, edgecolor='black')
axes[1].set_xticks(range(len(experiments)))
axes[1].set_xticklabels(experiments, rotation=15, ha='right')
axes[1].set_ylabel('PR-AUC Score', fontweight='bold')
axes[1].set_title('PR-AUC Comparison Across Experiments', fontsize=14, fontweight='bold')
axes[1].set_ylim([min(pr_values) - 0.05, 1.0])
axes[1].grid(True, alpha=0.3, axis='y')
for i, v in enumerate(pr_values):
    axes[1].text(i, v + 0.01, f'{v:.4f}', ha='center', va='bottom', fontweight='bold', fontsize=9)

plt.tight_layout()
plt.savefig('comparative_results.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "="*80)
print("ALL EXPERIMENTS COMPLETED SUCCESSFULLY!")
print("="*80)
print("\nKey Findings:")
print(f"✓ Best ROC-AUC: {results_summary['ROC-AUC'].max():.4f}")
print(f"✓ Best PR-AUC: {results_summary['PR-AUC'].max():.4f}")
print(f"✓ Most Robust: Experiment 5 (Cross-Validation)")
print("\nComparative visualization saved as 'comparative_results.png'")
```
