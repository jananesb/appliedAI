```python
known_faces = {}
known_names = []

folder_path = "known_faces"

for file in os.listdir(folder_path):
    
    # ✅ Allow only image files
    if not (file.endswith(".jpg") or file.endswith(".png") or file.endswith(".jpeg")):
        print("Skipping non-image file:", file)
        continue

    img_path = os.path.join(folder_path, file)

    img = cv2.imread(img_path)

    # ✅ Check if image loaded
    if img is None:
        print("Skipping invalid/corrupted image:", file)
        continue

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    face = extract_face(img)

    if face is not None:
        embedding = get_embedding(face)
        name = file.split('.')[0]

        known_faces[name] = embedding
        known_names.append(name)
    else:
        print("No face detected in:", file)

print("✅ Known faces stored:", known_names)
```
